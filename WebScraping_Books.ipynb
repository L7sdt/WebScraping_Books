```
#libs
from urllib.request import urlopen
from bs4 import BeautifulSoup
url1 = 'https://books.toscrape.com/catalogue/page-'
#https://books.toscrape.com/catalogue/page-1.html
url2 = '.html'

aux = []
liChildren = []
bookName = []
bookRating = []
bookPrice = []
k = -1

for page in range(1, 51):
  #print('Página ', page)
  html = urlopen(url1 + str(page) + url2)
  bs = BeautifulSoup(html, 'html.parser')
  items = bs.find_all('li', {'class': 'col-xs-6'})
  
  #loop through  the 'li' items
  for i in items:
    k+=1
    
    #getting books name
    title = i.findChild('h3')
    bookName.append(title.text)
    
    #getting books rating
    rate = i.find('p')
    aux.append(str(rate))
    if 'One' in aux[k]:
      bookRating.append('One')
    if 'Two' in aux[k]:
      bookRating.append('Two')
    if 'Three' in aux[k]:
      bookRating.append('Three')
    if 'Four' in aux[k]:
      bookRating.append('Four')
    if 'Five' in aux[k]:
      bookRating.append('Five')
      
    #getting books price
    price = i.find('p', {'class':'price_color'})
    bookPrice.append(price.text)
    
#using panda to create dataframe from lists above
import pandas as pd
df = pd.DataFrame({'name': bookName, 'rating': bookRating, 'price': bookPrice})

display(df)

```
